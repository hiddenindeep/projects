折线图：趋势随时间变化
柱状图：比较不同类别
散点图：观察变量之间的关系
箱线图：观察数据的分布情况
饼图：展示不同类别的占比

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

#折线图

#创建画布
plt.figure(figsize=(10,6))
rcParams['font.sans-serif'] = ['SimHei']

#数据
months = np.linspace(1,12,12)
sales = np.array([100,700,300,1000,500,1200,200,800,900,400,1100,600])

plt.plot(months,sales,color='green')
#设置标题
plt.title('月销售量',color='red')
#设置x轴和y轴标签
plt.xlabel('月份')
plt.ylabel('销售量')
#设置x轴刻度
plt.xticks(months)
#设置网格
plt.grid(True,linestyle='--',axis='y',linewidth=0.5)
#设置y轴范围
plt.ylim(0,1500)
#展示图例
plt.legend(['产品A'],loc='upper left')
#显示每个数据点值
for x,y in zip(months,sales):
    plt.text(x,y+3,str(y),ha='center',va='bottom')
#展示
plt.show()
```

```python
#饼图

#创建画布
plt.figure(figsize=(10,6))
rcParams['font.sans-serif'] = ['SimHei']

things = ['学习','娱乐','运动','睡觉','其它']
times = [6,4,2,8,4]

plt.pie(times,
        labels=things,
        autopct='%1.1f%%', #自动显示百分比
        startangle=0,  #起始角度
        wedgeprops={'width':0.5}, #扇形宽度
        explode=[0.1,0,0,0,0.1] #突出显示距离
        )
plt.title('一天时间分配',color='red')
plt.text(0,0,'总计:100%',ha='center',va='center')

#自动优化排版
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from matplotlib import rcParams

#散点图

#创建画布
plt.figure(figsize=(10,8))
rcParams['font.sans-serif'] = ['SimHei']

#数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0,10)
    x.append(tmp)
    y.append(tmp*2+random.gauss(0,2))

plt.scatter(x,y,color='blue',
            alpha=0.5,  #透明度
            s=10,       #点大小
            )
plt.title('X与Y的关系',color='red')
plt.xlabel('X自变量')
plt.ylabel('Y因变量')
plt.ylim(0,30)

#回归线
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')

#展示
plt.show()
```
